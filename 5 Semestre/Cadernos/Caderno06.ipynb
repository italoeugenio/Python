# Exercício 9

Encontre um arquivo de dados na internet, formato excel ou csv, faça a importação para um dataframe e sua análise preliminar.

```python
import pandas as pd
import os

arquivo = os.path.join('dados', '_Novas Tecnologias - Arrecadação por UF.csv')
dados = pd.read_csv(arquivo)

colunas_numericas = [
    'IMPOSTO SOBRE IMPORTAÇÃO', 'IMPOSTO SOBRE EXPORTAÇÃO', 'IPI - FUMO',
    'IPI - BEBIDAS', 'IPI - AUTOMÓVEIS', 'IPI - VINCULADO À IMPORTACAO',
    'IPI - OUTROS', 'IRPF', 'IRPJ - ENTIDADES FINANCEIRAS',
    'IRPJ - DEMAIS EMPRESAS', 'IRRF - RENDIMENTOS DO TRABALHO',
    'IRRF - RENDIMENTOS DO CAPITAL', 'IRRF - REMESSAS P/ EXTERIOR',
    'IRRF - OUTROS RENDIMENTOS', 'IMPOSTO S/ OPERAÇÕES FINANCEIRAS',
    'IMPOSTO TERRITORIAL RURAL', 'CPMF', 'COFINS'
]

colunas_numericas_ajustadas = [coluna for coluna in colunas_numericas if coluna in dados.columns]

for coluna in colunas_numericas_ajustadas:
    dados[coluna] = pd.to_numeric(dados[coluna].str.replace('.', '', regex=False).str.replace(',', '.', regex=False), errors='coerce')

dados_somados = dados.groupby('UF')[colunas_numericas_ajustadas].sum()

dados_somados_df = dados_somados.reset_index()

print(dados_somados_df)
```

```python
dados.head()
```

```python
dados.info()
```

```python
dados.describe()
```

```python
medias = dados[colunas_numericas].mean()

medias_formatadas = medias.apply(lambda x: f'{x:,.2f}')
print(medias_formatadas)
```

```python
import matplotlib.pyplot as plt

dados_somados_df['Total'] = dados_somados_df[colunas_numericas_ajustadas].sum(axis=1)

fig = plt.figure(figsize=(12, 12))
ax1 = fig.add_subplot()

ax1.pie(dados_somados_df['Total'], 
        labels=dados_somados_df['UF'], 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=90)

plt.title('Arrecadação Total por Estado')
plt.show()
```

```python
dados['Ano'] = pd.to_datetime(dados['Ano'], format='%Y')

estados_selecionados = ['RJ', 'SP', 'AM', 'DF']
dados_filtrados = dados[dados['UF'].isin(estados_selecionados)]

plt.figure(figsize=(12, 6))

for estado in estados_selecionados:
    filtro_estado = dados_filtrados['UF'] == estado
    plt.plot(dados_filtrados[filtro_estado]['Ano'], 
             dados_filtrados[filtro_estado]['IMPOSTO SOBRE IMPORTAÇÃO'],
             marker='o', 
             linestyle='-')

plt.xlabel('Ano')
plt.ylabel('Arrecadação')
plt.title('Arrecadação de Imposto sobre Importação por Estado ao Longo do Tempo')
plt.legend(estados_selecionados)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
